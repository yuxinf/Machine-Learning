```python
from sklearn.datasets import make_moons
from matplotlib import pyplot
from pandas import DataFrame
import pandas as pd
# generate 2d classification dataset
X, y = make_moons(n_samples=100, noise=0.1)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
from sklearn.datasets import make_circles
from matplotlib import pyplot
from pandas import DataFrame
# generate 2d classification dataset
X, y = make_circles(n_samples=100, noise=0.05, factor=.5)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
from sklearn.datasets import make_circles
from matplotlib import pyplot
from pandas import DataFrame
from sklearn import manifold, datasets

n_samples = 300
n_components = 2

# generate 2d classification dataset
X, y = make_circles(n_samples=100, noise=0.05 , factor=.4)

tsne = manifold.TSNE(n_components=n_components, init='random',
                         random_state=0, perplexity=10)
Y = tsne.fit_transform(X)

# scatter plot, dots colored by class value
#df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
df = DataFrame(dict(x=Y[:,0], y=Y[:,1], label=y))

colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
#dataset3 = Y.iloc[:, :].values
import pandas as pd
df3 = pd.DataFrame(Y)
df3.to_csv("dataset3.csv")
```
